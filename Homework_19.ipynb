## Гистограмма
### Histogram

```python
# Распределение баллов по уровню счастья в разных странах за 2019 год
df.Score.plot(kind='hist', 
             title='Распределение баллов по уровню счастья в разных странах за 2019 год',
             xlabel='Баллы',
             ylabel='Частота попадания')
```

Из полученных данных мы видим, что баллы по уровню счастья в разных странах за 2019 год чаще находится в среднем промежутке значений. Минимальные и максимальные значения баллов по уровню счастья отсутствуют. Счастье людей находится на "среднем" уровне.

```python
# Распределение ожидаемой продолжительности здоровой жизни в разных странах за 2019 год
df['Healthy life expectancy'].plot(kind='hist', 
             title='Распределение ожидаемой продолжительности здоровой жизни в разных странах за 2019 год',
             xlabel='Ожидаемая продолжительность здоровой жизни',
             ylabel='Частота попадания')
```

Из гистограммы "Распределение ожидаемой продолжительности здоровой жизни в разных странах за 2019 год" видим, что слишком низкие значения ожидаемой продолжительности здоровой жизни встречаются редко. Максимальные значения смещены в сторону увеличения значения. Распределение нормальное, асимметрия влево. 

## Ящик с усами, диаграмма размаха
### Box plot

```python
# Диаграмма размаха баллов по уровню счастья в разных странах за 2019 год
df.Score.plot(kind='box',
              title='Баллы по уровню счастья в разных странах за 2019 год')
```

Согласно диаграмме размаха баллов по уровню счастья в разных странах за 2019 год, выбросов данных В столбце с баллы по уровню счастья отсутствуют. Данные распределены нормально. Медианное значение баллов по уровню счастья находятся на среднем уровне. Отсутствуют слишком низкие и высокие быллы п уровню счастья.

## Точечная диаграмма
### Scatter Plot / диаграмма рассеяния

```python
# Построим взимосвязь баллов по уровню счастья в разных странах за 2019 год и ВВП на душу населения
df.plot(kind='scatter', 
        x='GDP per capita', 
        y='Score', 
        title='Взимосвязь баллов по уровню счастья в разных странах за 2019 год и ВВП на душу населения')
plt.show()
```

Их диаграммы рассеяния взимосвязи баллов по уровню счастья в разных странах за 2019 год и ВВП на душу населения видим, что данные показатели взаимосвязаны между собой. Связь прямая, чем выше ВВП на душу населения, тем выше вровень счастью в стране. Выбросы в указанных колонках, с большой сероятностью, отсутствуют.

## Тепловая карта
### Heat Map

```python
sns.heatmap(df.corr(method='kendall'))
```

Из полученной тепловой карты видим, что наиболее связаны значения Score и Overall rank. Взаимосвязь обратная. Чем выше баллы по уровню счастья в разных странах за 2019 год, тем меньше общая оценка. Иными словами, низкой общей оценке уровня счастья в разных странах за 2019 год соответствует высокие баллы. Это связано с тем, что на первом месте находится страна с высшим баллом по уровню счастья.

Сильно взаимосвязаны значения Score и GDP per capita, Social support , Healthy life expectancy. Взаимосвязь прямая. Чем больше ВВП на душу начеления, социальная поддержка и ожидаемая продолжительность здоровой жизни, тем больше выше баллы по уровню счастья.
