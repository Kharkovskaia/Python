Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш (см. справа)

Выгодна ли игра игроку?

Сколько в среднем приобретает
или теряет игрок за одну игру?

* 777: 200 руб.
* 999: 100 руб.
* 555: 50 руб.
* 333: 15 руб.
* 111: 10 руб.
* *77: 5 руб.
* **7: 3 руб.
* *00: 2 руб.
* **0: 1 руб.

'*'– любая цифра

Дополнительное задание повышенной сложности. Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

```python
import numpy as np
```

```python
trials = 10000000 # Количество попыток
amount = 0    # Сумма выигрыша 
for trial in range(trials):
    random_number = np.random.randint(0, 1000)    # Случайное число, выдаваемого лотерейным автоматом
    if random_number == 777:
        amount += 200
    elif random_number == 999:
        amount += 100
    elif random_number == 555:
        amount += 50
    elif random_number == 333:
        amount += 15
    elif random_number == 111:
        amount += 10
    elif (random_number % 100) == 77:
        amount += 5  
    elif (random_number % 10) == 7:
        amount += 3
    elif (random_number % 100) == 0:
        amount += 2  
    elif (random_number % 10) == 0:
        amount += 1
print('Средний выигрыш без учета стоимости игры равен', amount/trials)
print('Средний выигрыш с учетом стоимости игры равен', amount/trials - 1)
```

Проведем теоретический расчет среднего выигрыша (проигрыша).
Вероятность выпадения каждого из чисел 0,001.

* Вероятность получить 333 и выиграть 15 руб. - 0,001

* Вероятность получить **7 и выиграть 3 руб. - 0,1, без учета *77 - 0,09
* Вероятность получить *00 и выиграть 2 руб. - 0,01
* Вероятность получить **0 и выиграть 1 руб. - 0,1, без учета *00 - 0,09

Рассчитаем средний выигрыш (проигрыш) без учета стоимости билета

```python
0.001 * 200 + 0.001 * 100 + 0.001 * 50 + 0.001 * 15 + 0.001 * 10 + 0.009 * 5 + 0.09 * 3 + 0.01 * 2 + 0.09 * 1
```
